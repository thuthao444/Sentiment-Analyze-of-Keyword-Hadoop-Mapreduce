```python
import pandas as pd
path = "./Data/All_Comments_Final.csv"

df = pd.read_csv(path)
comments = df["Comments"]
comments
```

```python
import re

comments = [
    re.sub(r'\s+', ' ', re.sub(r'[^\x00-\x7F]', '', str(comment).strip()))  
    for comment in comments if isinstance(comment, str) and comment.strip()  
    if re.sub(r'[^\w]', '', comment.strip())
]

print("Cleaned comments:")
for i, c in enumerate(comments):
    print(i, c)
```

```python
import pandas as pd
import html

with open("./InputData/positive.txt", "r", encoding='latin-1') as pos_file:
    positive_words = set(pos_file.read().splitlines())

with open("./InputData/negative.txt", "r", encoding='latin-1') as neg_file:
    negative_words = set(neg_file.read().splitlines())

def calculate_score(comment):
    words = comment.lower().split()
    positive_count = sum(1 for word in words if word in positive_words)
    negative_count = sum(1 for word in words if word in negative_words)

    score = positive_count - negative_count

    return score

output = []

for index, comment in enumerate(comments):
    score = calculate_score(comment)
    post_id = index + 1
    user_id = 12345
    creation_date = "2011-09-16T08:00:53.557"

    escaped_text = html.escape(comment)

    row = f'<row Id="{post_id}" PostId="{post_id}" Score="{score}" Text="{escaped_text}" CreationDate="{creation_date}" UserId="{user_id}" />'
    output.append(row)

output_path = "./InputData/comments.txt"
with open(output_path, "w", encoding="utf-8") as file:
    file.write("\n".join(output))

print(f"File comments.txt đã được tạo tại: {output_path}")
```
